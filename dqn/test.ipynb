```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp_path = os.path.abspath('../../Tarea 3/policy_gradients')
```

```python
path_exp_csv_1 = exp_path + '\cartpole_exp_11\cartpole_exp_11_1.csv' 
path_exp_csv_2 = exp_path + '\cartpole_exp_11\cartpole_exp_11_2.csv'
path_exp_csv_3 = exp_path + '\cartpole_exp_11\cartpole_exp_11_3.csv' 
```

```python
df_1 = pd.read_csv(path_exp_csv_1, sep='\t').rename(columns={'avg_reward':'avg_reward_1', 'std_reward':'std_reward_1'})
df_1
```

```python
df_2 = pd.read_csv(path_exp_csv_2, sep='\t').rename(columns={'avg_reward':'avg_reward_2', 'std_reward':'std_reward_2'})
df_3 = pd.read_csv(path_exp_csv_3, sep='\t').rename(columns={'avg_reward':'avg_reward_3', 'std_reward':'std_reward_3'})
```

```python
test = pd.concat([df_1, df_2, df_3], axis=1)
test
```

```python
test['avg'] = test[['avg_reward_1', 'avg_reward_2', 'avg_reward_3']].mean(axis=1)
```

```python
test
```

```python
reward = test.loc[:, 'avg'].tolist()
total_steps = range(len(test))
plt.plot(total_steps, reward, marker='.',color='C0')
plt.ylabel('Average Reward')
plt.xlabel('Steps')
plt.grid('on')
# plt.pause(0.05)
# plt.savefig('enstimate_returns_pendulum_1.pdf')
# plt.close()
```
